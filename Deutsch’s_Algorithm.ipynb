Два входных кубита, инициализируются состояниями |0⟩ и |1⟩.

$$
|\psi_{0}\rangle = |01\rangle
$$

Затем к каждому кубиту применяется преобразование Адамара. Этот квантовый оператор приводит каждый из кубит в состояние суперпозиции, где существует равная вероятность измерения либо нуля, либо единицы. Его нормализованная унитарная матрица выглядит следующим образом:

$$
H = \frac{1}{\sqrt{2}}\begin{pmatrix}
1 & 1 \\
1 & - 1 \\
\end{pmatrix}
$$

Таким образом, кубиты теперь будут находиться в состоянии:

$$
\left| \psi_{1} \right\rangle = \left( \frac{\left| 0 \right\rangle + \left| 1 \right\rangle}{\sqrt{2}} \right)\left( \frac{\left| 0 \right\rangle - \left| 1 \right\rangle}{\sqrt{2}} \right)
$$

Следующим шагом в алгоритме является передача кубитов через унитарный оператор, который отображает |x, y⟩ → |x, y ⊕ f(x)⟩, где ⊕ обозначает XOR. Перед выполнением этой операции кубиты имеют вид:

$$
|\psi_{1}\rangle = \frac{\left| 00 \right\rangle - \left| 01 \right\rangle + \left| 10 \right\rangle - \left| 11 \right\rangle}{2}
$$

Прохождение через оракул:

* При f(0) = 0, f(1) = 0;

$$
|\psi_{2}\rangle = \frac{\left| 00 \right\rangle - \left| 01 \right\rangle + \left| 10 \right\rangle - \left| 11 \right\rangle}{2}
$$

$$
\left| \psi_{2} \right\rangle = \left( \frac{\left| 0 \right\rangle + \left| 1 \right\rangle}{\sqrt{2}} \right)\left( \frac{\left| 0 \right\rangle - \left| 1 \right\rangle}{\sqrt{2}} \right)
$$

* При f(0) = 1,f(1) = 1;


$$
|\psi_{2}\rangle = \frac{\left| 01 \right\rangle - \left| 00 \right\rangle + \left| 11 \right\rangle - \left| 10 \right\rangle}{2}
$$


$$
\left| \psi_{2} \right\rangle = \left( \frac{\left| 0 \right\rangle + \left| 1 \right\rangle}{\sqrt{2}} \right)\left( \frac{\left| 1 \right\rangle - \left| 0 \right\rangle}{\sqrt{2}} \right)
$$

$$
\left| \psi_{2} \right\rangle = - \left( \frac{\left| 0 \right\rangle + \left| 1 \right\rangle}{\sqrt{2}} \right)\left( \frac{\left| 0 \right\rangle - \left| 1 \right\rangle}{\sqrt{2}} \right)
$$

* При (f(0) = 0,f(1) = 1;

$$
|\psi_{2}\rangle = \frac{\left| 00 \right\rangle - \left| 01 \right\rangle + \left| 11 \right\rangle - \left| 10 \right\rangle}{2}
$$

$$
\left| \psi_{2} \right\rangle = \left( \frac{\left| 0 \right\rangle - \left| 1 \right\rangle}{\sqrt{2}} \right)\left( \frac{\left| 0 \right\rangle - \left| 1 \right\rangle}{\sqrt{2}} \right)
$$


* При f(0) = 1,f(1) = 0;


$$
|\psi_{2}\rangle = \frac{\left| 01 \right\rangle - \left| 00 \right\rangle + \left| 10 \right\rangle - \left| 11 \right\rangle}{2}
$$


$$
\left| \psi_{2} \right\rangle = \left( \frac{\left| 0 \right\rangle - \left| 1 \right\rangle}{\sqrt{2}} \right)\left( \frac{\left| 1 \right\rangle - \left| 0 \right\rangle}{\sqrt{2}} \right)
$$


$$
\left| \psi_{2} \right\rangle = - \left( \frac{\left| 0 \right\rangle - \left| 1 \right\rangle}{\sqrt{2}} \right)\left( \frac{\left| 0 \right\rangle - \left| 1 \right\rangle}{\sqrt{2}} \right)
$$

После прохождения оракула первый кубит проходит через гейт Адамара и происходит измерение его значения.

$$
\left| \psi_{3} \right\rangle = \left\{ \begin{matrix}
 \pm \left| 0 \right\rangle\left( \frac{\left| 0 \right\rangle - \left| 1 \right\rangle}{\sqrt{2}} \right),f(0) = f(1) \\
 \pm \left| 1 \right\rangle\left( \frac{\left| 0 \right\rangle - \left| 1 \right\rangle}{\sqrt{2}} \right),f(0) \neq f(1) \\
\end{matrix} \right.\ 
$$

```python
import random 
import cirq
from cirq import H, X, CNOT, measure
```

```python
def make_oracle(q0, q1, secret_numbers):

    if secret_numbers[0]:
        yield[CNOT(q0,q1), X(q1)] 

    if secret_numbers[1]:
        yield CNOT(q0, q1) 

def make_deutsch_circuit(q0, q1, oracle):
    c = cirq.Circuit()
    c.append([X(q1), H(q1), H(q0)]) 
    c.append(oracle) 
    c.append([H(q0), measure(q0, key='result')]) 
    
    return c
```

```python
q0, q1 = cirq.LineQubit.range(2) 

secret_numbers = [random.randint(0,1) for i in range(2)] 

oracle = make_oracle(q0, q1, secret_numbers) 
print('Secret function:\nf(x) = <{}>'.format(', '.join(str(e) for e in secret_numbers))) 

circuit = make_deutsch_circuit(q0,q1, oracle) 
print("Circuit:") 
print(circuit) 

simulator = cirq.Simulator() 
result = simulator.run(circuit)
print('Result of f(0)⊕f(1):') 
print(result) 
```
